REGプロット

```python
df['species'].value_counts()
```

```python
adelie = df.groupby('species').get_group('Adelie')
gentoo = df.groupby('species').get_group('Gentoo')
chinstrap = df.groupby('species').get_group('Chinstrap')
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=adelie, x='bill_length_mm', y='bill_depth_mm')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=gentoo, x='bill_length_mm', y='bill_depth_mm')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=chinstrap, x='bill_length_mm', y='bill_depth_mm')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.regplot(data=adelie, x='bill_length_mm', y='bill_depth_mm', scatter=False, label='adelie')
sns.regplot(data=gentoo, x='bill_length_mm', y='bill_depth_mm', scatter=False, label='gentoo')
sns.regplot(data=chinstrap, x='bill_length_mm', y='bill_depth_mm', scatter=False, label='chinstrap')
plt.legend()
plt.show()
```
