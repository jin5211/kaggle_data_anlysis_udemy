## 2. DataFrameからデータを抽出 

* カラムを指定して抽出
* locとilocでレコードとカラムを指定して抽出
* 条件文で抽出
* 複数の条件の場合の注意点
* filterを使う
* queryを使う

---

カラムを指定して抽出

```python
df.head(3)
```

```python
df[:3]
```

locとilocでレコードとカラムを指定して抽出  
※ loc : [index_label, column_label], iloc : [row_index, column_index]

```python
df.loc[0:3, 'Date':'Quantity']
```

```python
df.loc[[1, 3, 5], 'width']
```

```python
df.iloc[:3, :3]
```

```python
df.iloc[[0, 5, 10], [4, 5]]
```

```python
df.iloc[0:3, df.columns.get_loc('Color')]
# df.index.get_loc('ID001')
```

条件文で抽出

```python
df['height'] >= 9.5
```

```python
df[df['height'] >= 9.5]
```

複数の条件の場合の注意点

```python
# df[df['height']>= 9.5 & df['Price']>3000]
df[(df['height']>=9.5) & (df['Price']>3000)]
```

```python
df[(df['height']>=9.5) & (df['Price']>3000)|(df['Shape']=='triangle')]
```

```python
condition1 = df['height']>=9.5
condition2 = df['Price']>3000
condition3 = df['Shape']=='triangle'
df[condition1 & condition2 | condition3]
```

filterを使う  
※ locなどで対応できない場合に使用するという考えでよい  

```python
df.filter(like='or', axis=1).head()
```

```python
df.filter(like='0', axis=0).head()
```

queryを使う

```python
df.query('height > 9 & width < 3')
```

```python
df[(df['height']>9) & (df['width']<3)]
```

```python
df.query('Color in "red"')
```

```python
df.query('Color in ["red", "blue"]')
```

```python
df[df['Color'].isin(['red', 'blue'])]
```

## 6.重複、 欠損値の処理

* 重複レコードの削除
* 欠損値の確認
* 欠損値レコードを削除
* 欠損値を平均値で置換    
* scikit learn の SimpleImputerを使う
---

重複レコードの削除

```python
df_dup = df.loc[17:18]
df_dup
```

```python
# バージョンの更新によりdf = pd.concat([df, df_dup])が推奨される
df = pd.concat([df, df_dup])
df.shape
```

```python
df.tail()
```

```python
df.duplicated()
```

```python
df.duplicated()[16:20]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.tail()
```

欠損値の確認

```python
df.isnull().sum()
```

```python
df.isnull().sum(axis=1)
```

```python
df[df.isnull().sum(axis=1)>1]
```

欠損値レコードを削除

```python
# 欠損値の数が2以上のレコードを削除
# 1つ目の方法はidexを直接指定して削除
# df.drop(index=[4, 99])
df.dropna(thresh=9, inplace=True)
```

```python
df.shape
```

```python
df_not_null = df.dropna()
df_not_null.info()
```

```python
df.isnull().sum()
# これから欠損値を平均値で補完していく
# Date, Color, Shapeは保管できない
```

```python
# Date, Color, Shape に nanを持つレコードを削除
df.dropna(subset=['Date', 'Color', 'Shape'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

欠損値を平均値で置き換え

```python
hasnull_index = df[df.isnull().sum(axis=1)>0].index
hasnull_index
```

```python
null_check = df.isnull().sum()
null_check
```

```python
hasnull_cols = null_check[null_check.values!=0]
hasnull_cols
```

```python
hasnull_cols.index
```

```python
hasnull_cols = hasnull_cols.index.to_list()
hasnull_cols
```

```python
# df.loc[:, hasnull_cols].describe()
df.loc[:, hasnull_cols].mean()
```

```python
replace_dict = df.loc[:, hasnull_cols].mean().to_dict()
replace_dict
```

```python
replace_dict = {key: np.round(value, 2) for key, value in replace_dict.items()}
replace_dict
```

```python
df.fillna(replace_dict, inplace=True)
```

```python
hasnull_index = hasnull_index.to_list()
hasnull_index
```

```python
df.loc[hasnull_index]
```

scikit learn の SimpleImputer を使う

```python
df = pd.read_csv('../sec_2/sample_without_index.csv', usecols=['width', 'height'])
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
# インスタンスの作成
mean_imputer = SimpleImputer(strategy='mean')
```

```python
# データの学習
mean_imputer.fit(df[['width']])
```

```python
df['width_imputed'] = mean_imputer.transform(df[['width']])
df.head()
```

```python
df[df['width'].isnull()]
```

```python
# インスタンス化
median_imputer = SimpleImputer(strategy='median')
```

```python
# 学習
mean_imputer.fit(df[['height']])
```

```python
df['height_imputed'] = mean_imputer.transform(df[['height']])
df.head()
```

```python
df.agg(['mean', 'median'])
```

```python
df[(df['height'].isnull())|(df['width'].isnull())]
```
