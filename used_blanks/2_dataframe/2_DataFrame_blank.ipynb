平均50、標準偏差10の正規分布に従う小数データ (100, 1)からDataFrameを作成

```python
np.random.seed(17)
arr_nor = np.random.normal(50, 10, (100, 1))

df_score = pd.DataFrame(arr_nor, columns=['Score'])
df_score
```

['red', 'green', 'blue'] (100, 1)、 ['circle', 'square', 'triangle'] (100, 1)、ndarrayを結合しDataFrameを作成

```python
np.random.seed(17)
colors = ['red', 'green', 'blue']
shapes = ['circle', 'square', 'triangle']
arr_colors = np.random.choice(colors, (100, 1))
arr_shapes = np.random.choice(shapes, (100, 1))
```

```python
arr_cs = np.concatenate([arr_colors, arr_shapes], axis=1)
arr_cs[:5]
```

```python
arr_cs_hstack = np.hstack([arr_colors, arr_shapes])
arr_cs_hstack[:5]
```

```python
arr_cs_vstack = np.vstack([arr_colors, arr_shapes])
arr_cs_vstack[98:103]
```

```python
df_cs = pd.DataFrame(arr_cs, columns=['Color', 'Shape'])
df_cs
```

datetime 1997/7/5から1日刻み (100, 1)からDatetimeIndex、Seriesを作成

```python
dates = pd.date_range('1997/7/5', periods=100, freq='D')
# D: 日, B: 月曜~金曜, M: 月末, Y: 年末
# H: 時, T: 分, S: 秒
# 2D, 12Hなどとしても使用可
dates[:5]
# dataframe用のindexの型になっています
```

```python
# dataframe用のindexの型になっています
# dataframeのカラムとして扱いたいので、Seriesに変換しておきます
sr_date = pd.Series(dates, name='Date')
type(sr_date)
```

Series, DataFrameを結合

```python
df = pd.concat([sr_date, sr_price, sr_quantity, df_wh, df_quality, df_score, df_difference, df_cs], axis=1)
df
```

## 5.欠損値

* None, null, np.nanを確認
* DataFrameの欠損値はNoneもしくはnp.nanで置き換える
* df.isnull()の使い方
* 欠損値のある行を表示
* nullは平均値の計算などに影響するか
* 欠損値を視覚化
---

None, null, np.nanを確認

```python
print(None)
```

```python
# print(null)
# NameError: name 'null' is not defined
```

```python
print(np.nan)
```

DataFrameの欠損値はNoneもしくはnp.nanで置き換える

```python
null_values = [None, np.nan]
# dfの全体1000要素の内30個をnanで置き換えます
for i in range(30):
  np.random.seed(i)
  # index番号の生成(0~99)
  s = np.random.randint(0, 100)
  # column番号の生成(0~9)
  t = np.random.randint(0, 10)
  # null_valuesのindex番号を乱数で生成(0or1)
  u = np.random.randint(0, 2)
  df.iloc[s, t] = null_values[u]
```

```python
df.sample(5)
```

```python
df.sample(5)
```

```python
# NaT: Not a time
# NaN: Not a number
# None:....
```

df.isnull()の使い方

```python
df.isnull()
```

```python
# nullの数の確認(カラム毎)
df.isnull().sum()
```

```python
# df全体のnullの数
df.isnull().sum().sum()
```

```python
df.info()
```

```python
df.isnull().count()
# TrueもFalseも全部数えているため100となっている
```

```python
df.count()
# dfを直接countする場合は欠損値は含まれない
```

欠損値のある行を表示

```python
df.isnull()
```

```python
df.isnull().sum()
# sumはデフォルトでaxis=0になっているので、カラム方向の欠損値の数を数えて集計してくれます
```

```python
# axis=1 に変更してみる
df.isnull().sum(axis=1)
# sample(record)ごとのnullの数を集計してくれます
```

```python
df.isnull().sum(axis=1)>0
```

```python
df[np.random.choice([True, False], 100)]
```

```python
df[df.isnull().sum(axis=1)>0]
# 欠損値のある行を表示
```

```python
# Nullをハイライト
df[df.isnull().sum(axis=1)>0].style.highlight_null(color='red')
```

nullは平均値の計算などに影響するか

```python
df['Score'].mean()
```

```python
df['Score'].sum()
```

```python
df['Score'].sum()/df['Score'].mean()
```

```python
df['Score'].count()
```

```python
# 100個のScoreの値のうち、7個の値が欠損値となっています
```

欠損値の有無を視覚化  
※ seabornやグラフについての詳細は別のレクチャーで解説します

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), yticklabels=False)
plt.show()
```
